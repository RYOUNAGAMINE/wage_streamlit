```python
df_jp_ind = pd.read_csv('.\code_file\csv_data\雇用_医療福祉_一人当たり賃金_全国_全産業.csv',encoding='shift_jis')
df_jp_category = pd.read_csv('.\code_file\csv_data\雇用_医療福祉_一人当たり賃金_全国_大分類.csv',encoding='shift_jis')
df_pref_ind = pd.read_csv('.\code_file\csv_data\雇用_医療福祉_一人当たり賃金_都道府県_全産業.csv',encoding='shift_jis')
df_pref_category = pd.read_csv('.\code_file\csv_data\雇用_医療福祉_一人当たり賃金_都道府県_大分類.csv',encoding='shift_jis')
```

```python
df_temp = df_jp_category[df_jp_category['年齢']=='年齢計']
df_temp
```

```python
df_temp.groupby('産業大分類名')[['所定内給与額（万円）', '年間賞与その他特別給与額（万円）', '一人当たり賃金（万円）']].mean()
```

```python
df_pref_category.head()
```

```python
font_path = 'TakaoPGothic.ttf'
font_property = FontProperties(fname=font_path)
```

```python
age_list = df_pref_category['年齢'].unique()
print(age_list)
```

```python
df_pref_category['所定内給与額（万円）']=df_pref_category['所定内給与額（万円）'].replace('-', np.nan)
df_pref_category['年間賞与その他特別給与額（万円）']=df_pref_category['年間賞与その他特別給与額（万円）'].replace('-', np.nan)
df_pref_category['一人当たり賃金（万円）']=df_pref_category['一人当たり賃金（万円）'].replace('-', np.nan)
```

```python
df_pref_category.dropna(subset=['所定内給与額（万円）'], axis=0, inplace=True)
```

```python
age_list = df_pref_category['年齢'].unique()
print(age_list)
```

```python
df_pref_category = df_pref_category.astype({'所定内給与額（万円）': float, '年間賞与その他特別給与額（万円）': float, '一人当たり賃金（万円）': float})
wage_list =[]
for age in age_list:
  print(age)
  wage_temp = df_pref_category[df_pref_category['年齢'] == age]['一人当たり賃金（万円）'].values.tolist()
  wage_list.append(wage_temp)
```

```python
len(wage_list)
```

```python
fig = plt.figure(figsize=(16,6))
ax = plt.axes()
ax.set_title('年齢階級ごとの一人あたり賃金', fontproperties=font_property)
ax.set_xticklabels(age_list, fontproperties=font_property)
ax.boxplot(wage_list)
plt.show()
```

```python
fig = plt.figure(figsize=(16,6))
ax = plt.axes()
ax.set_title('年齢階級ごとの一人あたり賃金', fontproperties=font_property)
ax.set_xticklabels(age_list, fontproperties=font_property)
ax.boxplot(wage_list, showfliers=False)
plt.show()
```

```python
ind_list = df_pref_category['産業大分類名'].unique()
print(ind_list)
```

```python
wage_list =[]
for name in ind_list:
  print(name)
  wage_temp = df_pref_category[df_pref_category['産業大分類名'] == name]['一人当たり賃金（万円）'].values.tolist()
  wage_list.append(wage_temp)
```

```python
fig = plt.figure(figsize=(16,6))
ax = plt.axes()
ax.set_title('産業別の一人当たり賃金', fontproperties=font_property)
ax.set_xticklabels(ind_list, fontproperties=font_property, rotation=80)
ax.boxplot(wage_list)
plt.show()
```
